```python
"""
Food Delivery Analytics Pipeline
Created by: Mahmoud
Description: Cleaning and filtering restaurant sales data using NumPy and Pandas.
"""

import pandas as pd
import numpy as np

# 1. Define the original delivery dataset

delivery_data = {
    'OrderID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Item': ['Burger Combo', 'Pizza Large', 'Pasta Alfredo', 'Shawarma Wrap', 'Sushi Set', 
    'Burger Combo', 'Pizza Large', 'Salad Bowl', 'Shawarma Wrap', 'Sushi Set'],
    'Price': [150, 280, 190, 90, 450, 150, 280, 120, 90, 450],
    'Rating': [4.5, 3.8, 4.2, 4.7, 4.9, 4.0, 4.6, 3.5, 4.8, 2.1]
}
df = pd.DataFrame(delivery_data)
```

```python
# 2. Extract columns as NumPy arrays for vectorized operations
rating=df['Rating'].values  
prices=df['Price'].values
# Task 1: Calculate and print the average rating
print(f"⭐ Average Food Rating---> {rating.mean():.2f}")
```

```python
# Task 2: Find the cheapest meal using argmin(), set its price to 0, and print the updated DataFrame
df.loc[prices.argmin(), 'Price'] = 0
print("📊 Full Dataset After Modifying the Cheapest Meal:")
df
```

```python
# Task 3: Filter and print budget-friendly meals (Price < 150)
cheap_meals = df[df['Price'] < 150]
print("📌 Budget-Friendly Meals (Less than 150 EGP):")
print(cheap_meals)
```
